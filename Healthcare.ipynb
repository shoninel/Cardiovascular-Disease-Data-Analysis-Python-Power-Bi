```python
#machine learning
#seperate Independent & dependent Var
#create train set & test set
#Scaling
# Applying classification a;gorithm
#Evaluate the model
```

```python
data.head()
```

```python
#seperate Independent & dependent Var
X=data.drop(['target'],axis=1)
y=data['target']
```

```python
#create train set & test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
X_train.shape

y_train
#y_test
```

```python
X_test.shape
```

```python
X_train.head()
```

```python
#normalization/scale your data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#create the model
from sklearn.linear_model import LogisticRegression
log_red=LogisticRegression()
```

```python
log_red.fit(X_train,y_train)
```

```python
y_pred=log_red.predict(X_test)
```

```python
y_pred
```

```python
#evaluate machine learning model
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_pred,y_test))
```

```python
#print model accuracy & classfication report
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
```
